```r
va <- c(1:3,4,5)
vb <- 6:9
vc <- rep("c", times=6)
vd <- seq(1, 10, by=2)
```

```r
va
```

```r
vb
```

```r
vc
```

```r
rbind_abcd <- rbind(va, vb, vc, vd)
```

```r
rbind_abcd
```

```r
cbind_abcd <- cbind(va, vb, vc, vd)
cbind_abcd
```

```r
cbind_ab <- cbind(va, vb)
cbind_ab
```

```r
head(cbind_ab,2)
tail(cbind_ab,2)
print(apply(cbind_ab, 1, sum))
print(apply(cbind_ab, 2, sum))
print(apply(cbind_ab, 1, mean))
print(apply(cbind_ab, 2, mean))
```
