13. DF 데이터에서 "KIA", "SK", "LG" 팀의 year(년도)별 avg_mean(평균 타율)을
team_year_avg에 저장한 후 확인하세요. (Hint) year(년도), avg(타율), group_by(),
summarise(), filter()

```r
team_year_avg <-
  DF %>%
    filter(team %in% c("KIA", "SK", "LG")) %>%
    group_by(team, year) %>%
    summarise(avg_mean = mean(avg, na.rm = TRUE))

team_year_avg %>% show()
```

14. team_year_avg에서 "KIA"팀의 정보는 kia, "SK"팀의 정보는 sk, "LG"팀의 정보는 lg에 저장
하세요.

```r
kia <- team_year_avg %>% filter(team == 'KIA')

sk <- team_year_avg %>% filter(team == 'SK')

lg <-team_year_avg %>% filter(team == 'LG')
```

15. 아래의 스크립트가 의미하는 것이 무엇인지 설명하고 그래프를 그려 확인하세요

```r
# point와 line 함수
plot(x = team_year_avg$year, y = team_year_avg$avg_mean, type = "n", main = "Teambatting average by Year", xlab = "YEAR", ylab = "AVG")
# LG
points(x = lg$year, y = lg$avg_mean, col = "purple", pch = 8, cex = 1)
lines(x = lg$year, y = lg$avg_mean, col = "purple")
# SK
points(x = sk$year, y = sk$avg_mean, col = "green", pch = 11, cex = 1)
lines(x = sk$year, y = sk$avg_mean, col = "green")
# KIA
points(x = kia$year, y = kia$avg_mean, col = "orange", pch = 4, cex = 1)
lines(x = kia$year, y = kia$avg_mean, col = "orange")
legend("bottomright", legend = c("LG","SK","KIA"), pch = c(8,11,4), col = c("purple","green", "orange"), cex = 1.5)
```

19. DF 데이터에서 team(팀)별 R(득점)의 합, H(안타)의 합을 team_R_H_sum에 저장하고 확인
하세요.

```r
team_R_H_sum <-
  DF %>%
    group_by(team) %>%
    summarise(R_sum = sum(R), H_sum = sum(H))

team_R_H_sum
```

20. team_R_H_sum 데이터가 아래와 같이 표현되도록 스크립트를 작성하세요.

```r
team_R_H_sum %>%
  ggplot(aes(x = team, y = R_sum)) +
  geom_bar(stat = "identity", aes(reorder(team,-R_sum), fill = team))
```

21. team_R_H_sum 데이터가 아래와 같이 표현되도록 스크립트를 작성하세요.

```r
team_R_H_sum %>%
  ggplot(aes(x = team, y = R_sum)) +
  geom_bar(stat = "identity", aes(reorder(team,-R_sum), fill = team)) +
  coord_flip() +
  labs(x = "team")
```
