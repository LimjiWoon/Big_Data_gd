13번. dplyr패키지를 이용하여 DF에서 team(소속팀)과 batter_name(타자 이름)을 group으로 묶어
소속팀의 타자별 평균 G(출전게임수), 평균 AB(타수(타석-볼넷-사구-희생번트-희생플라
이)), 합계 R(득점)을 출력하세요.

```r
DF %>% group_by(team, batter_name) %>% summarise(G_AVG = mean(G), AB_AVG = mean(AB), R_SUM = sum(R), .groups = "drop") %>% show()
#, .groups = "drop"를 안넣으면 정보가 남아있다고 경고가 뜬다.

#더 안정적으로 짜면 이렇게 된다.
DF %>% group_by(team, batter_name) %>%
  summarise(G_AVG = mean(G, na.rm = T), AB_AVG = mean(AB, na.rm = T), R_SUM = sum(R, na.rm = T), .groups = "drop") %>% show()
```

21번 (숙제) dplyr패키지를 이용하여 DF에서 비복원 추출 방식으로 30개의 샘플을 추출해 train_set에 저
장하고 출력하세요. (Hint) sample_n(), replace=F 비복원추출

```r
#sample_n은 사용자가 지정한 수만큼 랜덤하게 데이터를 가져온다
train_set <- DF %>% sample_n(size = 30, replace = F)
train_set %>% show()
```

22번 (숙제)  dplyr패키지를 이용하여 DF에서 비복원 추출 방식으로 70%의 샘플을 추출해 train_set에 저
장하고 출력하세요. (Hint) sample_frac()

```r
#sample_frac은 사용자가 지정한 비율만큼 랜덤하게 데이터를 가져온다
train_set <- DF %>% sample_frac(size = 0.7, replace = F)
train_set %>% show()
```

23번 (숙제) . dplyr패키지를 이용하여 DF에서 위 문제에서 구한 train_set을 제외한 샘플을 추출하여
test_set에 저장하고 출력하세요. (Hint) setdiff()

```r
#setdiff(DF1, DF2)로 DF2와 곂치지 않는 DF1의 값들을 반환한다.
test_set <- setdiff(DF, train_set)
test_set %>% show()
```
