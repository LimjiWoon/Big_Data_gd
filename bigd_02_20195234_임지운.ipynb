17번 (20195234 임지운)

```r
temp <- c(sample(1:10, 20, replace = T), -15:11, 21:25, NA)
#A temp에 저장된 데이터의 갯수 출력
temp
length(temp)
#B temp에 저장된 NA의 갯수 출력
table(is.na(temp))
#C 이상값을 NA로 변환하여 temp_clear_outlier에 저장
summary(temp)
iqr <- IQR(temp, na.rm = T) #중간에 결측값이 존재하므로 IQR이 있으면 에러가 발생한다. 결측값을 무시하고 계산하게 한다.
lowerfence <- summary(temp)[2] - iqr*1.5 #Q1 - iqr * 1.5
upperfence <- summary(temp)[5] + iqr*1.5 #Q3 + iqr * 1.5
#ifelse를 이용하여 이상값 범위에 있는 값들은 NA로 아니라면 기존값을 넣어준다.
temp_clear_outlier <- ifelse((temp < lowerfence | temp > upperfence), NA, temp)
#D temp_clear_outlier에 저장된 NA의 갯수를 출력
table(is.na(temp_clear_outlier))
#E temp_clear_outlier에서 NA를 제거하여 temp_new에 저장
temp_new <- na.omit(temp_clear_outlier)
#F temp_new에 저장된 데이터 출력
temp_new
```
