14. train_set 데이터를 이용하여 sales와 관련된 youtube, facebook, newspaper의 다중 선형 회
기 모델 yfn_model의 회귀식을 구하고 결과를 해석하세요.

```r
print("yfn_model = sales ~ youtube + facebook + newspaper")
yfn_model <- lm(formula = sales ~ youtube + facebook + newspaper, data = train_set)
summary(yfn_model)
```

## <결과의 해석>
**ufn_model**


1.   회귀식: sales = 3.544772 + 0.046076*youtube + 0.187840*facebook
2.   제일 밑에 F-statistic의 p-value값은 2.2e-16 로 0.05보다 작기에 이 회귀식은 회귀분석 모델 전체에 대해 통꼐적으로 의미가 있다고 볼 수 있다.
3.   중간의 Coefficients:에는 y절편 값(Intercept) 및 변수들의 p-value값이 나와있다. newspaper 변수의 경우 NA로 sales를 설명하는데 유의하지 않다.
4.   밑 부분의 Adjusted R-squared:  0.9098으로 91%만큼의 설명력을 가진다고 판단할 수 있다.
